```julia
using LazySets, LinearAlgebra, Random

```

```julia
function random_zonotope(dimensions::Int, order::Int; rng=Random.default_rng())
    c = randn(rng, dimensions)               # random center
    m = dimensions * order                   # number of generators
    G = randn(rng, dimensions, m) ./ sqrt(dimensions)  # scaled random generators
    return Zonotope(c, G)
end

# 5D zonotope with order 3
Z = random_zonotope(5, 3)
```
